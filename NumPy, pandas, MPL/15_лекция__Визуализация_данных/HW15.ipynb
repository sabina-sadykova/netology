```python
from pylab import rcParams
rcParams['figure.figsize'] = 10,7

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns
```

```python
df = pd.read_csv('netflix_titles.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['date_added', 'rating'])
```

```python
#Изменим формат дат
df["date_added"] = pd.to_datetime(df['date_added'])
df['day_added'] = df['date_added'].dt.day
df['year_added'] = df['date_added'].dt.year
df['month_added']=df['date_added'].dt.month
df['year_added'].astype(int);
df['day_added'].astype(int);
```

```python
df.head()
```

```python
#Сравним количество фильмов и шоу
sns.countplot(df['type'])
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.title('Type')
```

Фильмов в выборке значительно больше, чем шоу

```python
#Распределение выходов фильмов и шоу по годам

df_year=df.groupby(["year_added", "type"])["title"].count()
df_year.head()
```

```python
df_year.unstack().plot(kind='bar')
```

Количество выпущенных новых фильмов резко выросло в 2016 году

```python
#Зависимость между рейтингом и типом контента
plt.figure(figsize=(10,8))
sns.countplot(x='rating',hue='type',data=df)
```

Наибольшое количество фильмов имеет рейтинги TV и R

```python
#Распределение по самым популярным жанрам фильмов
plt.style.use("ggplot")
plt.figure(figsize=(12,6))
df[df["type"]=="Movie"]["listed_in"].value_counts()[:20].plot(kind="barh")
```

Самые популярные жанры - Комедии и Драмы

```python
#Анализ выхода фильмов и шоу по месяцам и годам
df_date = df.pivot_table(
                        index='month_added', 
                        columns='year_added', 
                        values='title', 
                        aggfunc=lambda x: len(x.unique())).fillna(0).applymap(float)

df_date
```

```python
ax = sns.heatmap(df_date)
```

Больше всего фильмов вышло в ноябре 2019 года

```python
import re
df['dur_min'] = df['duration'].str.extract('(\d+)').astype(int) 
```
