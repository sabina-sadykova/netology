### 10.Проверить несколько гипотез для данных, чтобы понять почему сотрудники покидают компанию.

#### H0: Средняя удовлетворенность покинувших компанию сотрудников и работающих одинаковая

Перед использоватение тестов на равенство средних проверим на нормальность распределение переменной satisfaction_level

```python
stat, p = scipy.stats.shapiro(df['satisfaction_level']) # тест Шапиро-Уилк
print('Statistics=%.3f, p-value=%.3f' % (stat, p))
alpha = 0.05
if p > alpha:
    print('Принять гипотезу о нормальности')
else:
    print('Отклонить гипотезу о нормальности')
```

Распределение переменной не соответсвует нормальному, использовать t-критерий нельзя, для проверки равества средних воспользуемся критерием Манна-Уитни.

```python
import matplotlib
matplotlib.use('TkAgg')
from scipy import stats

u, p_value = stats.mannwhitneyu(d0['satisfaction_level'], d1['satisfaction_level'])
print("two-sample wilcoxon-test", p_value)
```

р_value <= 0.05 как следствие нулевая гипотеза отвергается и мы можем говорить о том, что средние различны

Сделаем визуальную проверку

```python
plt.hist(d0['satisfaction_level'], bins=24, color='g', alpha=0.75)
plt.hist(d1['satisfaction_level'], bins=24, color='y', alpha=0.55)
plt.show()
```

```python
x0 = statistics.mean(d0['satisfaction_level'])
x1 = statistics.mean(d1['satisfaction_level'])
print('Среднее по покинувшим компанию:', x0)
print('Среднее по оставшимся в компании:', x1)
```

Уровень удовлетворенности по текущим работникам значительно выше, чем по уволившимся.
Таким обзразом, на увольнение также оказывает внияние уровень удовлетворенности

#### Н0: Среднее по количеству повышений у покинувших компанию сотрудников совпадает с оставшимися

Воспользуемся тестом Манна-Уитни для проверки гипотезы

```python
import matplotlib
matplotlib.use('TkAgg')
from scipy import stats

u, p_value = stats.mannwhitneyu(d0['promotion_last_5years'], d1['promotion_last_5years'])
print("two-sample wilcoxon-test", p_value)
```

р_value <= 0.05 как следствие нулевая гипотеза отвергается и мы можем говорить о том, что средние различны

Сделаем визуальную проверку

```python
d2 = df.groupby(['left','promotion_last_5years'])['satisfaction_level'].agg(['count'])
d3 = d2.groupby(level=0).apply(lambda x:
                                                 100 * x / float(x.sum()))
d3
```

```python
d3.unstack().plot.barh(stacked=True)
```

Доля получивших повышение за последние 5 лет среди уволившихся сотрудников значительно меньше, чем среди текущих сотрудников
